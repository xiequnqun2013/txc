```python
user_click = click.join(user.set_index('user_id'), on='user_id')
```

```python
user_click_ad = user_click.join(ad.set_index('creative_id'), on='creative_id')
```

```python
group_user_click_ad = user_click_ad.groupby('user_id').agg(lambda x: '%'.join(x.values.astype(str)))
```

```python
group_user_click_ad.to_csv('group_user_click_ad.csv')
```

```python
def genOneHot(row,maxIndex):
    ins = row.strip().split('%')
    vectors =  [0]*maxIndex
    for i in ins:
        if  i == '\\N' :
                continue
        vectors[int(i)] = 1
    return vectors
```

```python
group_user_click_ad['low_ids'] = group_user_click_ad['product_category']
group_user_click_ad['low_ids'] = group_user_click_ad['low_ids'].astype('object')
# 第一次先抽取部分特征来进行计算
for index, row in group_user_click_ad.iterrows():
    product_category_vector =  genOneHot(row['product_category'],100)
    industry_vector =  genOneHot(row['industry'],1000)
    time_vector =  genOneHot(row['time'],100)
    vectors = product_category_vector + industry_vector + time_vector
    #print (type(vectors),vectors)

    group_user_click_ad.at[index,'low_ids'] = vectors
```

```python
train_data = group_user_click_ad[['low_ids']]
train_data
```

```python
train_data_sample = train_data.join(user.set_index('user_id'), on='user_id')
```
