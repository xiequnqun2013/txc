```python
# gen fastText sample
import pandas as pd
group_user_click_ad = pd.read_csv('/home/tione/notebook/group_user_click_ad.csv')
```

```python
# only use creative_id
user = pd.read_csv('/home/tione/notebook/data/train_preliminary/user.csv')
traing_data = group_user_click_ad[['user_id','creative_id','time','ad_id','product_id','product_category',
                                   'advertiser_id','industry']].join(user.set_index('user_id'), on='user_id')
```

```python
traing_data
```

```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        pass
 
    try:
        import unicodedata
        unicodedata.numeric(s)
        return True
    except (TypeError, ValueError):
        pass
 
    return False
replace_traing_data = traing_data.applymap(lambda x: x if is_number(x) else x.replace('%',' ')) # lambda x,y : x%y if x>y else x**y
```

```python
replace_traing_data
```

```python
replace_traing_data['training_features'] = replace_traing_data['creative_id'].apply(lambda x: "c ".join(x.split(" "))+"c") + ' ' + replace_traing_data['time'].apply(lambda x: "t ".join(x.split(" "))+"t") +' '+ replace_traing_data['ad_id'].apply(lambda x: "a ".join(x.split(" "))+"a")  +' '+ replace_traing_data['product_id'].apply(lambda x: "pr ".join(x.split(" "))+"pr") + ' '+ replace_traing_data['product_category'].apply(lambda x: "pc ".join(x.split(" "))+"pc") + ' '+ replace_traing_data['advertiser_id'].apply(lambda x: "ai ".join(x.split(" "))+"ai") +' '+ replace_traing_data['industry'].apply(lambda x: "in ".join(x.split(" "))+"in")
```

```python
replace_traing_data['training_features']
```

```python
replace_traing_data['age_training'] = '__label__' + replace_traing_data['age'].astype(str) + ' '+ replace_traing_data['training_features']
```

```python
replace_traing_data['age_training']
```

```python
replace_traing_data['age_training'].to_csv('/home/tione/notebook/fastText-0.9.2/data/full_age_training.cvs',index=False,header=False)
```

```python
replace_traing_data['gender_training'] = '__label__' + traing_data['gender'].astype(str) + ' '+ replace_traing_data['training_features']
```
